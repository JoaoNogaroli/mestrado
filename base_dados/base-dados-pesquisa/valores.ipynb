```python
import pandas as pd
```

```python
df = pd.read_csv('bpa-2025.csv', sep=';', encoding='latin-1')
```

```python
df.columns
```

```python
lista_ativos = df['DENOM_CIA'].unique().tolist()
lista_final = []
for ativo in lista_ativos:
    dfativo = df[df['DENOM_CIA']==ativo]
    datas = dfativo['DT_REFER'].unique().tolist() 
    caixa = dfativo.query('DT_FIM_EXERC==@datas and CD_CONTA=="1.01.01"')['VL_CONTA'].values
    valor_ativo = dfativo.query('DT_FIM_EXERC==@datas and CD_CONTA=="1"')['VL_CONTA'].values
    if len(caixa) == 3 and len(valor_ativo) == 3:
        dicio = {
            'ativo': ativo,
            'caixas':{
                datas[0]: caixa[0],
                datas[1]: caixa[1],
                datas[2]: caixa[2]
            },
            'valores_ativo': {
                datas[0]: valor_ativo[0],
                datas[1]: valor_ativo[1],
                datas[2]: valor_ativo[2],
            }
        }
        lista_final.append(dicio)


```

```python
pd.DataFrame(lista_final)
```
