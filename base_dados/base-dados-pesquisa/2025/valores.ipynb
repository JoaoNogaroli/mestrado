## -------- INICIO DAS FORMULAS (MAGIC FORMULA E PIOTROSKI)

```python
df_final.head()
```

```python
import pandas as pd
import numpy as np


DATAS = ['2025-03-31', '2025-06-30', '2025-09-30']
resultados = []

for _, row in df_final.iterrows():
    empresa = row['ativo']
    
    for data in DATAS:
        try:
            # Contas do período
            caixa        = row[f'caixa_{data}']
            divida_cp    = row[f'divida_cp_{data}']
            divida_lp    = row[f'divida_lp_{data}']
            pl           = row[f'pl_{data}']
            ebit         = row[f'ebit_{data}']
            lucro        = row[f'lucro_liquido_{data}']
            receita      = row[f'receita_liquida_{data}']
            ativo_total  = row[f'ativo_total_{data}']

            divida_bruta    = divida_cp + divida_lp
            divida_liquida  = divida_bruta - caixa
            cap_investido   = pl + divida_liquida

            # ---- Indicadores calculáveis sem preço ----
            roic         = ebit / cap_investido   if cap_investido != 0  else np.nan
            roe          = lucro / pl             if pl != 0             else np.nan
            margem       = lucro / receita        if receita != 0        else np.nan
            div_ebitda   = divida_bruta / ebit    if ebit != 0           else np.nan  # proxy EBITDA≈EBIT por ora
            roa          = lucro / ativo_total    if ativo_total != 0    else np.nan

            resultados.append({
                'ativo':        empresa,
                'dt_refer':     data,
                'ebit':         ebit,
                'cap_investido': cap_investido,
                'ROIC':         roic,
                'ROE':          roe,
                'margem_liq':   margem,
                'div_ebitda':   div_ebitda,
                'ROA':          roa,
                # EV e Earnings Yield entram depois com o preço
                'EV':           np.nan,
                'earnings_yield': np.nan,
            })

        except Exception as e:
            print(f"Erro em {empresa} | {data}: {e}")

df_ind = pd.DataFrame(resultados)

# Remove cap_investido negativo (distorce ROIC — PL negativo)
df_ind = df_ind[df_ind['cap_investido'] > 0]

print(df_ind.shape)
print(df_ind[['ativo','dt_refer','ROIC','ROE','margem_liq']].head(10).to_string(index=False))
```
