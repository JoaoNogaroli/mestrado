```python
import os
from brapi import Brapi
from dotenv import load_dotenv
import pandas as pd
import requests
import json

from pathlib import Path
```

```python
lista_df = []
for filename in os.listdir('estatistica'):
    
    # 2. Reconstruct the full absolute or relative path to the file
    try:
        full_path_balanco = os.path.join('balanco', filename)
        full_path_dre = os.path.join('dre', filename)
        full_path_estatistica = os.path.join('estatistica', filename)

        df_balanco = pd.read_csv(full_path_balanco)
        df_dre = pd.read_csv(full_path_dre)
        df_estatistica = pd.read_csv(full_path_estatistica)

        df_merge = df_balanco.merge(df_dre, on='endDate').merge(df_estatistica, on='endDate').copy()

        df_merge['capital_investido'] = df_merge['totalCurrentAssets'].fillna(0) - df_merge['currentLiabilities'].fillna(0) + df_merge['propertyPlantEquipment'].fillna(0)
        df_merge['ROC'] = df_merge['ebit'] / df_merge['capital_investido']
        df_merge['EY']  = df_merge['ebit'] / df_merge['enterpriseValue']


        dicio_merge = {
            'ativo':filename,
            'data':df_merge
        }
        lista_df.append(dicio_merge)
    except Exception as e:
        print(e)
        continue
```

```python

lista_final = []
retirar_ativos = []
for i in range(len(lista_df)):
    if len(lista_df[i]['data'][['endDate','capital_investido','ROC','EY']]) >= 10:
        ativo = lista_df[i]['ativo']
        data_f = lista_df[i]['data'][['endDate','ROC','EY']]
        data_f.to_csv(f'magic_formula/{ativo}')
        dc = {
            'ativo':lista_df[i]['ativo'],
            'data':data_f
        }
        lista_final.append(dc)
    else:
        try:
            # path_remove = os.path.join('magic_formula', lista_df[i]['ativo'])
            # os.remove(path_remove)
            retirar_ativos.append(lista_df[i]['ativo'])
            print(f'{lista_df[i]['ativo']} - Tamanho: {len(lista_df[i]['data'][['endDate','capital_investido','ROC','EY']])}')
            print('ATIVO ESCLUIDO da MF: ',lista_df[i]['ativo'])
        except:
            print('erro ao excluir')
            continue
```
