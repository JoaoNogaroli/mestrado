```python

dataframe_collection = []
ativos_problema = []

for stock in lista_ativos:
    # ticker = stock.stock
    try:    
        ticker = stock

        dataframe_stock = []

        url = f"https://brapi.dev/api/v2/stocks/statistics?symbols={ticker}&startDate=2012-12-31&endDate=2025-12-31&mode=history"
        print(url)
        response = requests.request("GET", url, headers = {
        "Authorization": api_k
        })

        
            
        
        # print(response.json()['results'])
        info_final = {
            'ativo':stock,
            'data':pd.DataFrame(response.json()['results'][0]['data']).set_index('endDate')
        }
        # print(info_final)
        dataframe_stock.append(info_final)

        dataframe_collection.append(dataframe_stock)
    except Exception as e:
        print(e)
        print("Ticker com problema: ",ticker)
        ativos_problema.append(ticker)

        continue
```

```python
for ativo in dataframe_collection:
    nome_ativo = ativo[0]['ativo']
    df = ativo[0]['data']
    df.to_csv('estatistica/'+nome_ativo)
```
