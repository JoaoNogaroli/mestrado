## DADOS DE ENTRADA

```python
valor_total_investir = 50000
total_ativos = 20
tamanho_carteira = 5
limite_minimo_erro = float(-0.002)
limite_maximo_erro = float(0.05)
```

## INICIANDO MODELO

```python
model = pyo.ConcreteModel()
```

#### DEFININDO VARIAVEIS

```python
## Quantidade de lotes que vao ser comprados
model.x = pyo.Var(range(5), bounds=(0,0.15))
x = model.x
```

```python
## se o ativo foi usado ou nao 0 nao 1 sim
model.y = pyo.Var(range(total_ativos), bounds=(0,1))
y=model.y
```

#### DEFININDO RESTRIÇÔES

```python
model.quantidade_ativos_carteira = pyo.Constraint(expr = sum(y[i] for i in range(total_ativos)) == tamanho_carteira)
```

```python
model.valor_maximo_investido = pyo.Constraint(expr = sum(x[i]*100*preco_inicial_ativo[i] for i in range(tamanho_carteira)) <= valor_total_investir)
```

```python
model.soma_peso_percent = pyo.Constraint(expr = sum(x[i] for i in range(int(tamanho_carteira))) == 1 )
```
